```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def laplacian_operator(image):
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Apply the Laplacian operator
    laplacian = cv2.Laplacian(gray_image, cv2.CV_64F)
    
    # Normalize the result to display properly
    laplacian = cv2.normalize(laplacian, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_8U)
    
    return laplacian

def display_laplacian(original_image, laplacian_image):
    fig, axs = plt.subplots(1, 2, figsize=(10, 5))
    
    # Display original image
    axs[0].imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
    axs[0].set_title('Original Image')
    axs[0].axis('off')
    
    # Display Laplacian image
    axs[1].imshow(laplacian_image, cmap='gray')
    axs[1].set_title('Laplacian Image')
    axs[1].axis('off')
    
    plt.show()

# Path to the directory containing the images
images_directory = "C:\\Users\\kadal\\Desktop\\images"

# Iterate through each image in the directory
for i in range(1, 4):  # Assuming you have images named from 1.png to 3.png
    image_path = os.path.join(images_directory, f"{i}.png")
    
    # Check if the file exists
    if os.path.isfile(image_path):
        original_image = cv2.imread(image_path)
        
        # Apply Laplacian operator
        laplacian_image = laplacian_operator(original_image)
        
        # Display the results
        display_laplacian(original_image, laplacian_image)
    else:
        print(f"File not found: {image_path}")
```
