```python
import numpy as np
import cv2
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import os

def kmeans_segmentation(image, n_clusters):
    pixels = image.reshape((-1, 3))
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(pixels)
    segmented_image = kmeans.labels_.reshape(image.shape[:2])
    return segmented_image
```

```python
def display_segmentation(original_image, segmented_image):
    fig, axs = plt.subplots(1, 2, figsize=(10, 5))
    
    # Display original image
    axs[0].imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
    axs[0].set_title('Original Image')
    axs[0].axis('off')
    
    # Display segmented image
    axs[1].imshow(segmented_image, cmap='viridis', vmin=0, vmax=n_clusters - 1)
    axs[1].set_title('Segmented Image')
    axs[1].axis('off')
    
    plt.show()
```

```python
# Path to the directory containing the images
images_directory = "C:\\Users\\kadal\\Desktop\\images"

# Number of clusters for K-means
n_clusters = 3

# Iterate through each image in the directory
for i in range(1, 4):  # Assuming you have images named from 1.png to 10.png
    image_path = os.path.join(images_directory, f"{i}.png")
    
    # Check if the file exists
    if os.path.isfile(image_path):
        original_image = cv2.imread(image_path)
        
        # Perform K-means segmentation
        segmented_image = kmeans_segmentation(original_image, n_clusters)
        
        # Display the results
        display_segmentation(original_image, segmented_image)
    else:
        print(f"File not found: {image_path}")
```
